# Practice Problem - Temperatures

Use the [data/munich_temperatures_average.txt](data/munich_temperatures_average.txt) data file, which gives the temperature in Munich every day for several years. Read in the data, and print out the minimum, average maximum temperature for each year, e.g:

    1995: -3C  10C  35C
    1996: ...
    
You can also try and print out the same values, but for each month (averaged over years), e.g.:

    January: -15C 0C 3C
    February: ...

## Solution

```python
# your solution here
import numpy as np
import math

#SOLUTION 1
f = open('data/munich_temperatures_average.txt', 'r')

old_year = 0
default_min = 1000
default_max = -1000
min_temp = default_min
max_temp = default_max
tot_temp = 0
ntot = 0
for line in f:
    line = line.strip()
    columns = line.split()
    new_year = math.floor(float(columns[0]))
    temp = float(columns[1])
    if new_year != old_year and old_year != 0:
        print(str(old_year)+': '+str(min_temp)+'C '+str(tot_temp/ntot)+'C '+str(max_temp)+'C')
        min_temp = default_min
        max_temp = default_max
        tot_temp = 0
        ntot = 0
    old_year = new_year
    if temp < min_temp:
        min_temp = temp
    if temp > max_temp:
        max_temp = temp
    tot_temp += temp
    ntot += 1
print(str(old_year)+': '+str(min_temp)+'C '+str(tot_temp/ntot)+'C '+str(max_temp)+'C')
f.close()

print()

#SOLUTION 2
year = []
temp = []
with open('data/munich_temperatures_average.txt', 'r') as f:
    for line in f:
        y,t = line.strip().split()
        year.append(float(y))
        temp.append(float(t))
    ymin = int(min(year))
    ymax = int(max(year))
    for i in range(ymin,ymax+1):
        days = [i < j < i+1 for j in year]
        minday = days.index(True)
        maxday = minday + days.count(True)
        print(str(i)+': '
              +str(min(temp[minday:maxday]))+'C '
              +str(np.mean(temp[minday:maxday]))+'C '
              +str(max(temp[minday:maxday]))+'C')

print()

#SOLUTION 3
d = {}
with open('data/munich_temperatures_average.txt', 'r') as f:
    for line in f:
        year,temp = line.strip().split()
        y = year[0:4]
        if y not in d:
            d[y]=[]
        d[y].append(float(temp))

for key in sorted(d):
    temps = d[key]
    print(key+':', min(temps), sum(temps)/len(temps), max(temps))

print()

#NOW FOR EACH MONTH
d = {}
with open('data/munich_temperatures_average.txt', 'r') as f:
    for line in f:
        year,temp = line.strip().split()
        y = year[0:4]
        frac = float(year)-float(y)
        m = int(12.*frac)+1
        if m not in d:
            d[m]=[]
        d[m].append(float(temp))

for key in sorted(d):
    temps = d[key]
    print(str(key)+':', min(temps), sum(temps)/len(temps), max(temps))
```
